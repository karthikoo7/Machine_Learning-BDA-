# X and Y Split
target = fitnss level

```python
X = df.drop(columns=['Fitness_Level'])
Y = df['Fitness_Level']

X.shape,Y.shape
```

```python

Y.unique()
```

```python
label_map = {'Excellent':3, 'Good':2, 'Average':1, 'Poor':0}

Y.replace(label_map,inplace=True)
```

```python
Y.unique()
```

```python
X_ohe = pd.get_dummies(X)
X_ohe.shape
```

```python
cat_col = X_ohe.columns[X_ohe.nunique() <= 15]
cat_col.shape
```

```python
cont_col = X_ohe.columns[X_ohe.nunique() > 15]
cont_col.shape
```

Train Test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X_ohe,Y,test_size=0.3,random_state=7,stratify=Y)

X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

applying standard scaler on continous & outlier detection

```python
from sklearn.preprocessing import RobustScaler
```

```python
rc = RobustScaler()
X_train[cont_col] = rc.fit_transform(X_train[cont_col])
X_test[cont_col] = rc.transform(X_test[cont_col])
```

Outlier Imputation



```python
def outlier_imputation_iqr(df,col):
  df1 = df.copy()
  q1,q3 = df1[col].quantile([0.25,0.75])
  iqr = q3-q1
  LB = q1 - 1.5*iqr
  UB = q3 + 1.5*iqr
  print("Column -->",col)
  df1.loc[(df1[col]<LB),col] = LB
  df1.loc[(df1[col]>UB),col] = UB
  print(LB,UB, df1[col].min(),df1[col].max())
  return df1
```

```python
X_train_io = X_train.copy()
for col in cont_col:
  X_train_io = outlier_imputation_iqr(X_train_io,col)
```

RFE - Recursive Feature Selection

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
rfe = RFE(LogisticRegression(random_state=7),verbose=2)
rfe.fit(X_train,Y_train)


for col,rank in zip(X_train.columns,rfe.ranking_):
  print(col,rank)
```

```python
selected_col = X_train.columns[rfe.ranking_ == 1]
selected_col.shape
```

```python
X_train_rfe = X_train.loc[:,selected_col]
X_test_rfe = X_test.loc[:,selected_col]
```
