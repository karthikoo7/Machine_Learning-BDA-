```python
import pandas as pd
df = pd.read_csv("sample_1.csv")
df.shape
```

```python
X = df.loc[:,["c1","c2"]]
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test = train_test_split(X,
                                  test_size=0.3,
                                  random_state=7)
X_train.shape,X_test.shape
```

```python
from sklearn.cluster import KMeans
mykmeans = KMeans(3,random_state=7)
mykmeans.fit(X_train)
y_pred = mykmeans.predict(X_test)
```

```python
# Plot
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=df.iloc[:,2], cmap='viridis')

plt.title("3 Clusters with Small Noise")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.show()
```

```python
inertia_values = []
for k in range(2,11):
    mykmeans = KMeans(k,random_state=7)
    mykmeans.fit(X_train)
    inertia_values.append(mykmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(list(range(2,11)), inertia_values)
```

# Evaluation Parameters

```python
from sklearn.metrics import silhouette_score
```

```python
# two parameters to silhouette_score
# X : X_test data
# labels : cluster numbers predicted by the algorithm
y_pred = mykmeans.predict(X_test)
silhouette_score(X_test, y_pred)
```
