label encoding -- done for ordinal data -- matains order like -0,1,2,3.
one hot encoding (dummy variables) -- for nominal columns : this avoids introdcution of order in data


```python
# label Encoding
label_map = {'BusinessTravel':{'Travel_Rarely':1,'Travel_Frequently':2,'Non-Travel':0}}
df.replace(label_map,inplace=True)
df['BusinessTravel'].unique()
```

```python
df.dtypes
```

```python
# OHE - one hot encoding
df_ohe = pd.get_dummies(df)
df_ohe.shape,df.shape
```
