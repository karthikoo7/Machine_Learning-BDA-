Task: on Heart disease dataset

Load the dataset in a pandas dataframe
Q1. Rename column V1 to "Serial No"
Q2. Drop column V2
Q3. In column V5, V6, V7, V8, and V9 replace -9 with np.nan (NA)
Q4. In V12 column replace -9 value with mean of this column
Q5. Replace -9 in all columns with np.nan(NA)
Q6. Count number of NA in every column
Q7. Drop columns where percentage of NA is greater than 90%
Q8. Find columns where number of unique values are less than or equal to 10.
For example, V4 has two unique values 0 and 1.
Q9. Find columns where number of unique values are greater than 10.
For example, V3 has many unique values.
Q10. Find how many values in V58 are 0, 1 and 2.
Q11. In column V58 convert all 2s in 1s.
Q12. Save  this data frame as a new csv file



```python
data = pd.read_csv('76_attributes_heartdiseases.csv')
```

```python
print(data.shape)
data.isna().sum()
```

```python
data.dtypes
```

```python
# Q1
data.rename(columns={"V1":"Serial No"},inplace=True)
data.columns
```

```python
# Q2
data.drop(columns='V2',inplace=True)
data.columns
```

```python
# Q3
col = ['V5', 'V6', 'V7', 'V8', 'V9']
data[col] = data[col].replace(-9,np.nan)

data.head()
```

```python
# Q4
data['V9'] = data['V9'].replace(-9,data['V9'].mean())
data.head()
```

```python
# Q5
data['V75'] = data['V75'].replace('name',-9)
data['V75'] = data['V75'].astype(float)
data['V75'] = data['V75'].replace(-9,np.nan)

data = data.replace(-9,np.nan)
data
```

```python
# Q6
data.isna().sum()
```

```python
# 7
list_cols = data.columns[data.isna().sum()/len(data) * 100 > 90]
```

```python
data.drop(columns=list_cols,inplace=True)
```

```python
data.shape
```

```python
# Q8
list_cols = data.columns[data.nunique() <= 10]
list_cols
```

```python
# Q9
list_great = data.columns[data.nunique()>10]
list_great
```

```python
# Q10
data['V58'].value_counts()
```

```python
# Q11
data['V58'] = data['V58'].replace(2,1)
data['V58']
```

```python
# 12
data.to_csv('76_attributes_heartdiseasesData.csv')
```
